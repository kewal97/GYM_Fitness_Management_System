```python
# create a Tkinter window
window = tk.Tk()
window.title('SQL QUERY RESULT')

# create a Canvas widget with a fixed size
canvas = tk.Canvas(window)
canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

# create a vertical scrollbar for the canvas
scrollbar = tk.Scrollbar(window, orient=tk.VERTICAL)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

# configure the canvas to use the scrollbar
canvas.configure(yscrollcommand=scrollbar.set)
canvas.bind('<Configure>', lambda e: canvas.configure(scrollregion=canvas.bbox("all")))

# configure the scrollbar to scroll the canvas
scrollbar.config(command=canvas.yview)

# create a Frame widget inside the Canvas widget to hold the content
frame = tk.Frame(canvas)

label = tk.Label(frame, text='Enter a query number(1 to 7):')
entry = tk.Entry(frame)
button = tk.Button(frame, text='Submit', command=queries, bg='green')
label.pack()
entry.pack()
button.pack()

# Create a window to hold the Canvas and Scrollbar
canvas.create_window((0, 0), window=frame, anchor="nw")

# update the scrollbar to match the canvas's height
frame.bind('<Configure>', lambda e: canvas.config(scrollregion=canvas.bbox("all")))
    
window.mainloop()
```
